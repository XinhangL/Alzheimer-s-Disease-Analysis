### EDA

```python
# shape and data types of the data
print(df.shape)
print(df.info())

# select numeric columns
df_numeric = df.select_dtypes(include=[np.number])
numeric_cols = df_numeric.columns.values
print(numeric_cols)

# select non numeric columns
df_non_numeric = df.select_dtypes(exclude=[np.number])
non_numeric_cols = df_non_numeric.columns.values
print(non_numeric_cols)
```

```python
df.head()
```

**Data cleansing**

```python
# % of missing.
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))

cols = df.columns[:30] # first 30 columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
# Replace the Missing
df['USAN Stem'] = df['USAN Stem'].fillna('00None0')
df['ATC Codes'] = df['ATC Codes'].fillna('None')
df['Level 1 ATC Codes'] = df['Level 1 ATC Codes'].fillna('None')
df['Level 2 ATC Codes'] = df['Level 2 ATC Codes'].fillna('None')
df['Level 3 ATC Codes'] = df['Level 3 ATC Codes'].fillna('None')
df['Level 4 ATC Codes'] = df['Level 4 ATC Codes'].fillna('None')
```

```python
# drop unimportant Feature
cols_to_drop = ['Parent Molecule ChEMBL ID.1', 'Parent Molecule Type', 'Target Type', 'Target Organism', 'Smiles', 'Binding Site Name', 'Binding Site Comment', 'Mechanism Comment', 'Selectivity Comment', 'References', 'USAN Stem', 'Synonyms']
df = df.drop(cols_to_drop, axis=1)
df.head()
```

```python
# check for duplicate
duplicate = df[df.duplicated()]
duplicate.count()
```

```python
# check outliers
sns.boxplot(data=df)
```

```python
# replace unnecessary data
df['Parent Molecule ChEMBL ID'] = df['Parent Molecule ChEMBL ID'].apply(lambda x:x[6:]).tolist()
df['Target ChEMBL ID'] = df['Target ChEMBL ID'].apply(lambda x:x[6:]).tolist()
df.head()
```
